```python
#import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", palette="muted")
from numbers import Number
from scipy import stats
#allow plots and visualisations to be displayed in the report
%pylab inline
```

```python
# Read weather data into pandas dataframe
# Data taken from https://www.ncdc.noaa.gov/data-access/land-based-station-data/land-based-datasets/global-historical-climatology-network-ghcn


df=pd.read_csv('2017 simple weather data.csv')

# Data info
df.info()
```

```python
df.head()
```

CODES FOR MEASURES \
    TMAX = Maximum temperature (tenths of degrees C) \
    TMIN = Minimum temperature (tenths of degrees C) \
    TAVG = Average temperature (tenths of degrees C) \
    TSUN = Daily total sunshine (minutes)

```python
sunshine_df = df.loc[df['measure'] == 'TSUN']
sunshine_df
```

```python
sunshine_df.describe()
```

```python
sunshine_df.groupby('Country')[['data']].mean()/24
```

```python
tavg_df = df.loc[df['measure'] == 'TAVG']
tavg_df_group = tavg_df.groupby('Country')[['data']].mean()/10
tavg_df_group
```

```python
tmin_df = df.loc[df['measure'] == 'TMIN']
tmin_df_group = tmin_df.groupby('Country')[['data']].mean()/10
tmin_df_group
```

```python
tmax_df = df.loc[df['measure'] == 'TMAX']
tmax_df_group = tmax_df.groupby('Country')[['data']].mean()/10
tmax_df_group
```

```python
weather_df = pd.merge(tmax_df_group, tmin_df_group, left_on= 'Country', right_on='Country')
weather_df
```

```python
weather_df = pd.merge(weather_df, tavg_df_group, left_on= 'Country', right_on='Country')
weather_df
```

```python
weather_df.columns = ['max', 'min','aver']
weather_df
```

```python
weather_df.to_csv ('average country temperature.csv', index = True, header=True)
```
