```python
import pandas as pd
import numpy as np 
import matplotlib as plt
```

```python
master = pd.read_csv('master.csv')
master
```

```python
bad_countries = ['Dominica',
'Grenada',
'Jamaica',
'Kiribati',
'Saint Kitts and Nevis',
'Saint Lucia',
'Saint Vincent and Grenadines',
'San Marino',
'Seychelles',
'Albania',
'Antigua and Barbuda',
'Armenia',
'Aruba',
'Azerbaijan',
'Bahamas',
'Bahrain',
'Barbados',
'Belize',
'Bulgaria',
'Costa Rica',
'Cyprus',
'Macau',
'Maldives',
'Malta',
'Mongolia',
'Montenegro',
'Nicaragua',
'Oman']
```

```python
master_good = master[~master['country'].isin(bad_countries)]
```

If you want to get the combined single suicide rate/100k per country per year.

```python
combined_rate_per_year = master_good.groupby(['year','country'])['suicides/100k pop'].mean()
combined_rate_per_year
```

If you want to add multiple columns, use agg

```python
combined_rate_gdp = master_good.groupby(['year','country']).agg({'suicides/100k pop': 'mean', 'gdp_per_capita ($)': 'mean'})
combined_rate_gdp
```
